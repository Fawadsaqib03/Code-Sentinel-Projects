# ***Cleaning The Data***

```python
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
Columns_to_drop = [
    "zero", "zero.1", "zero.2", "zero.3", "zero.4", "zero.5", "zero.6",
    "zero.7", "zero.8", "zero.9", "zero.10", "zero.11", "zero.12",
    "zero.13", "zero.14", "zero.15", "zero.16", "zero.17", "zero.18"
]

# Drop columns
df = df.drop(columns=Columns_to_drop, errors="ignore")
df
```

```python
# Filling Missing Values in Embarked Column
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].median())
```

```python
# Rename column 'Pclass' â†’ 'Class'
df = df.rename(columns={"Pclass": "Class"})

# Rename column '2urvived' â†’ 'Survived'
df = df.rename(columns={"2urvived": "Survived"})

# Check remaining columns
print(df.columns)
```

```python
df.isnull().sum()
```

# ***Regression Analysis***

```python
# Heatmap - Correlation of numerical features
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap of Numeric Features", fontsize=14)
plt.show()
```

```python
# Bar Plot - Survival count by Gender
plt.figure(figsize=(8,6))
sns.countplot(x="Sex", hue="Survived", data=df)
plt.title("Survival Count by Gender", fontsize=14)
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Bar Plot - Survival count by Passenger Class
plt.figure(figsize=(8,6))
sns.countplot(x="Class", hue="Survived", data=df)
plt.title("Survival Count by Passenger Class", fontsize=14)
plt.xlabel("Passenger Class")
plt.ylabel("Count")
plt.show()
```

```python
# Pie Chart - Class distribution
class_counts = df["class"].value_counts()
plt.figure(figsize=(6,6))
plt.pie(class_counts, labels=class_counts.index, autopct="%1.1f%%", startangle=90)
plt.title("Passenger Class Distribution")
plt.show()
```

```python
# Histogram - Fare distribution
plt.figure(figsize=(8,6))
sns.histplot(df["Fare"].dropna(), bins=40, kde=True)
plt.title("Fare Distribution of Passengers", fontsize=14)
plt.xlabel("Fare")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter Plot - Age vs Fare
plt.figure(figsize=(8,6))
sns.scatterplot(x="Age", y="Fare", hue="Survived", data=df, alpha=0.7)
plt.title("Age vs Fare (Colored by Survival)", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```

```python
# Histogram - Age distribution
plt.figure(figsize=(8,6))
sns.histplot(df["age"].dropna(), bins=30, kde=True)
plt.title("Age Distribution of Passengers", fontsize=14)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```
